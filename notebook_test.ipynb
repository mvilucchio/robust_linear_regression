```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_file(eps, delta1, delta2):
    return f"data/reg_param optimal/Huber double noise - eps {eps} - reg_param optimal - alphas [0.01 100 36] - delta [{delta1} {delta2}].npz"

fig, ax = plt.subplots(1, 1, figsize=(10, 8), tight_layout=True)

eps = 0.3
delta1 = 0.1
delta2_list = [0.5, 1.0, 2.0, 5.0, 10.0]
for delta2 in delta2_list:
    data = np.load(find_file(eps, delta1, delta2))
    ax.plot(data['alphas'], data['errors'], label = f"$\Delta$ = [{delta1}, {delta2}]")


ax.set_ylabel(r"$\frac{1}{d} E[||\hat{w} - w^\star||^2]$")
ax.set_xlabel(r"$\alpha$")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_title(f"$\epsilon$ = {eps}, a = {1.0}")
ax.minorticks_on()
ax.grid(True, which="both")
ax.legend()
```
